```python
import numpy as np
from PIL import Image
```

```python
Image.open("grid.png")
```

```python
image = Image.open("grid.png").convert("RGBA")
```

```python
data = np.array(image)
```

```python
data.shape
```

```python
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        r,g,b,a = data[i,j];

        data[i,j,3] = 255 if r*g*b>0 else 0 
```

```python
Image.fromarray(data)
```

```python
Image.fromarray(data).save("grid.png")
```

```python


for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        r,g,b,a = data[i,j];
        
        if not a or r|g|b==0:
          data[i,j] = [18, 20, 34,a]
        else:
           data[i,j] = [205, 247, 255,a]
```

```python
Image.fromarray(data)
```

```python
Image.fromarray(data).save("grid.png")
```
